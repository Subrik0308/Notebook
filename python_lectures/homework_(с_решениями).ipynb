```python
vvod = int(input())  # Решение через while
result = 1
if vvod % 2 == 0:
    i = 2
    while i <= vvod:
        result = result * i
        i += 2
else:
    i = 1
    while i <= vvod:
        result = result * i
        i += 2
print(result)
```

Напишите программу, которая принимает на вход строку, и выводит слово, которое встречается во фразе чаще всего. Гарантируется, что такое слово будет только одно. Регистр слов не учитывается, знаки препинания в предложениях игнорируются.

**Ввод**

    В единственную строку вводится строка.

**Вывод**

    В единственную строку выводится искомое слово и то количество раз, сколько оно встретилось в тексте.
    Данные выводятся через запятую.

**Пример**

Ввод

    Напишите программу, которая принимает на вход строку, и выводит слово, которое встречается во фразе чаще всего. Гарантируется, что такое слово будет только одно. Регистр слов не учитывается, знаки препинания в предложениях игнорируются.

Вывод

    слово, 2

```python
import string
from collections import Counter

stroka = input()

cleaned_list = []
for word in stroka.split():
    cleaned_list.append(word.strip(string.punctuation).lower())

result = Counter(cleaned_list).most_common(1)

for i in result:
    print(*i, sep = ", ")
```

```python
# Альтернативный способ обработки вывода результата

import string
from collections import Counter

stroka = input()

cleaned_stroka = []
for word in stroka.split():
    cleaned_stroka.append(word.strip(string.punctuation).lower())

result = Counter(cleaned_stroka).most_common(1)

print('{}, {}'.format(result[0][0], result[0][1]))
```

**Пример 2**

    Иванов Иван Иванович, 25.01.1980
    Петров Иван Иванович, 17.03.2006
    Сидоров Иван Иванович, 16.03.2006

Вывод

    Петров И.И. может участвовать в выборах

```python
# Сырой код, хоть и рабочий, в рамках текущей задачи

import datetime

string = (list(map(str,input().split(', '))))

voter_list = []
for i in range(int(string[0])):
    voter_list.append(list(map(str,input().split(', '))))

voting_date = datetime.datetime.strptime(string[1], "%d.%m.%Y")

for i in voter_list:
    full_name = (i[0]).split(' ')
    birthday = datetime.datetime.strptime(i[1], "%d.%m.%Y")

    age = voting_date.year - birthday.year - ((voting_date.month, voting_date.day) < (birthday.month, birthday.day))

    abbreviated_name = (full_name[0] + ' ' + full_name[1][0] + '.' + full_name[2][0] + '.')

    if age >= 18:
        print(f'{abbreviated_name} может участвовать в выборах')
    else:
        print(f'{abbreviated_name} НЕ может участвовать в выборах')
```

```python
# Доработанная версия версия решения с комментариями

import datetime

def get_date(string_date):
    # Преобразование данных типа "строка", в тип "дата"
    formatted_date = datetime.datetime.strptime(string_date, "%d.%m.%Y")
    return formatted_date

def аbbreviation(full_name_str):
    # Получаем из полного ФИО, сокращённый вариант
    full_name_split = full_name_str.split(' ')
    reduced_name = full_name_split[0] + ' ' + full_name_split[1][0] + '.' + full_name_split[2][0] + '.'
    return reduced_name

string = input().split(', ')  # Ввод количества избирателей и даты голосования в формате «3, 16.03.2024»

voting_date = get_date(string[1])  # Преобразование даты голосования из строки в тип "дата"

voter_list = []  # Новый список, куда будем собирать ФИО и даты рождения трёх избирателей
for i in range(int(string[0])):  # string[0] указывает, данных скольких участников будут введены и помещены в список
    voter_list.append(input().split(', '))

voter_list_result = []  # Новый список, куда будем собирать результаты for
for voter in voter_list:
    birthday = get_date(voter[1])  # Преобразование даты рождения избирателя из строки в тип "дата"
    
    # Проверяем, есть ли избирателю полные 18 лет на дату голосования
    if (voting_date.year - birthday.year - ((voting_date.month, voting_date.day) < (birthday.month, birthday.day))) >= 18:
        get_voter_list = f"{аbbreviation(voter[0])} может участвовать в выборах"
    else:
        get_voter_list = f"{аbbreviation(voter[0])} НЕ может участвовать в выборах"
    voter_list_result.append(get_voter_list)  # Собираем результаты в итоговый список

print(*voter_list_result, sep="\n")  # Распаковываем итоговый список и выводим результат каждого избирателя на новой строке
```

## Задача 5

Напишите простой калькулятор, который считывает с пользовательского ввода три строки: первое число, второе число и операцию, после чего применяет операцию к введённым числам ("первое число" "операция" "второе число") и выводит результат на экран (в консоль).

Поддерживаемые операции: +, -, /, *, mod, pow, div, где

    mod — это взятие остатка от деления,
    pow — возведение в степень,
    div — целочисленное деление.

Если выполняется деление и второе число равно 0, необходимо выводить строку "Деление на 0!".

Если операция не определена, то выводится сообщение «Ошибка ввода».

Обратите внимание, что на вход программе приходят вещественные числа.

**Ввод**

    В первою строку вводится первое число.
    Во вторую строку вводится второе число.
    В третью строку вводится операция.

**Вывод**

    В единственную строку выводится результат.

**Пример 1**

Ввод

    5
    5
    +

Вывод

    10.0

**Пример 2**

Ввод

    5
    0
    /

Вывод

    Деление на 0!

**Пример 3**

Ввод

    5
    5
    5

Вывод

    Ошибка ввода

```python
def calculator(number1, number2, operator):
    
    number1, number2 = float(number1), float(number2)
    
    if operator == "+":
        result = number1 + number2
    elif operator == "-":
        result = number1 - number2
    elif operator == "*":
        result = number1 * number2
    elif operator == "/":
        if number2 == 0:
            result = "Деление на 0!"
        else:
            result = number1 / number2
    elif operator == "mod":
        if number2 == 0:
            result = "Деление на 0!"
        else:
            result = number1 % number2
    elif operator == "pow":
        result = number1 ** number2
    elif operator == "div":
        if number2 == 0:
            result = "Деление на 0!"
        else:
            result = number1 // number2
    else:
        result = "Ошибка ввода"
    
    return(result)

number1 = input()
number2 = input()
operator = input()

print(calculator(number1, number2, operator))
```

## Задача 6

Дано целое число N > 0. Если N — нечетное, то вывести произведение 1·3·...·N; если N — четное, то вывести произведение 2·4·...·N.

**Ввод**

В единственную строку вводится целочисленное число больше нуля.

**Вывод**

В единственную строку выводится результат произведения.

```python
vvod = int(input())  # Решение через while
result = 1
if vvod % 2 == 0:
    i = 2
    while i <= vvod:
        result = result * i
        i += 2
else:
    i = 1
    while i <= vvod:
        result = result * i
        i += 2
print(result)
```

```python
vvod = int(input())  # Решение через range
result = 1
for i in range(2 - (vvod % 2), vvod + 1, 2):
    result *= i
print(result)
```

Например, если дан список [12, 104, 81], то сумма всех его цифр будет равна:

1 + 2 + 1 + 0 + 4 + 8 + 1 = 17

```python
def sum_st(num):
    num = str(num)
    sum_num = 0
    
    for i in num:
        sum_num += int(i)
    
    return sum_num

spisok = list(map(int,input().split(',')))
split = str()
for i in spisok:
    if i < 0:
        i = str(abs(i))
        split += i
    else:
        i = str(i)
        split += i

print(sum_st(split))
```

Далее в каждой строке вводятся элементы строки списка, через пробел. Каждая новая строка списка должна вводится с новой строки.

В единственную строку через пробел выводится сумма элементов столбцов, сначала сумму элементов нулевого столбца, затем первого, второго и т.д.

```python
strings, columns = map(int, input().split())

spisok = []
for i in range(strings):
    spisok.append(list(map(int,input().split())))

summ = [sum(i) for i in zip(*spisok)]
print(*summ)
```

Напишите программу, которая принимает на вход строку, и выводит слово, которое встречается во фразе чаще всего. Гарантируется, что такое слово будет только одно. Регистр слов не учитывается, знаки препинания в предложениях игнорируются.

**Ввод**

В единственную строку вводится строка.

**Вывод**

В единственную строку выводится искомое слово и то количество раз, сколько оно встретилось в тексте. Данные выводятся через запятую.

**Пример**

Ввод

Напишите программу, которая принимает на вход строку, и выводит слово, которое встречается во фразе чаще всего. Гарантируется, что такое слово будет только одно. Регистр слов не учитывается, знаки препинания в предложениях игнорируются.

Вывод

слово, 2

```python
import string
from collections import Counter

stroka = input()

cleaned_list = []
for word in stroka.split():
    cleaned_list.append(word.strip(string.punctuation).lower())

result = Counter(cleaned_list).most_common(1)

for i in result:
    print(*i, sep = ", ")
```

```python
# Альтернативный способ обработки вывода результата

import string
from collections import Counter

stroka = input()

cleaned_stroka = []
for word in stroka.split():
    cleaned_stroka.append(word.strip(string.punctuation).lower())

result = Counter(cleaned_stroka).most_common(1)

print('{}, {}'.format(result[0][0], result[0][1]))
```

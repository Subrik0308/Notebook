Задано два числа A и B. Необходимо вывести символ "<", если A < B, ">" , если A> B и "=", если A = B

**Ввод**

В первой строке задается значение первого числа. Во второй строке задается значение второго числа.

**Вывод**

Ввод: </br>
100 </br>
200

Вывод: </br>
<

```python
a = int(input())
b = int(input())

if a < b:
    print('<')
elif a > b:
    print('>')
else:
    print('=')
```

**Ввод**

В первой строке задается зарплата первого сотрудника. </br>
Во второй строке задается зарплата второго сотрудника. </br>
В третей строке задается зарплата третьего сотрудника.

**Вывод**

В единственную строку выводится разница между максимальной и минимальной зарплатами.

**Пример**

Ввод: </br>
100 </br>
200 </br>
300

Вывод: </br>
200

```python
rab1 = int(input())
rab2 = int(input())
rab3 = int(input())
if rab1 >= rab2 and rab1 >= rab3:
    max = rab1
elif rab2 >= rab1 and rab2 >= rab3:
    max = rab2
else:
    max = rab3
if rab1 <= rab2 and rab1 <= rab3:
    min = rab1
elif rab2 <= rab1 and rab2 <= rab3:
    min = rab2
else:
    min = rab3

print(max - min)
```

## Задача 5

Вводится целое число. Если оно четырехзначное, то используя операции целочисленного деления и остаток от деления, найти сумму и произведение его цифр. Если число не четырехзначное, то вывести сообщение «Ошибка ввода».

**Ввод**

В единственную строку вводится целочисленное число больше нуля.

**Вывод**

В первой строке выводится результат суммы, согласно шаблону: </br>
Сумма числа 1234 равна 10 </br>
Во второй строке выводится результат произведения, согласно шаблону: </br>
Произведение числа 1234 равна 24

Если условие задачи не соответствует условию, т.е. число не четырёхзначное, то в единственной строке выводится «Ошибка ввода».

**Пример 1**

Ввод: </br>
1234

Вывод: </br>
Сумма числа 1234 равна 10 </br>
Произведение числа 1234 равна 24

**Пример 2**

Ввод: </br>
123 

Вывод: </br>
Ошибка ввода

```python
vvod = int(input())
if vvod // 1000 != 0 and vvod // 1000 < 10 and vvod // 1000 > -10:
    n0 = vvod % 10
    n1 = vvod // 10 % 10
    n2 = vvod // 100 % 10
    n3 = vvod // 1000
    summ = n0 + n1 + n2 + n3
    multiplication = n0 * n1 * n2 * n3
    print(f'Сумма числа {vvod} равна {summ}')
    print(f'Произведение числа {vvod} равна {multiplication}')
else:
    print('Ошибка ввода')
```

## Задача 6

Дано целое число N > 0. Если N — нечетное, то вывести произведение 1·3·...·N; если N — четное, то вывести произведение 2·4·...·N.

**Ввод**

В единственную строку вводится целочисленное число больше нуля.

**Вывод**

В единственную строку выводится результат произведения.

```python
vvod = int(input())  # Решение через while
result = 1
if vvod % 2 == 0:
    i = 2
    while i <= vvod:
        result = result * i
        i += 2
else:
    i = 1
    while i <= vvod:
        result = result * i
        i += 2
print(result)
```

```python
vvod = int(input())  # Решение через range
result = 1
for i in range(2 - (vvod % 2), vvod + 1, 2):
    result *= i
print(result)
```
